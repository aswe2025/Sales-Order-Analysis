```python
import pandas as pd
import sqlalchemy as sa
from urllib.parse import quote_plus

# 1. Connection Setup
params = quote_plus(
    r"DRIVER={ODBC Driver 17 for SQL Server};"
    r"SERVER=localhost\SQLEXPRESS;"
    "DATABASE=Sales;"
    "Trusted_Connection=yes;"
)
engine = sa.create_engine(f"mssql+pyodbc:///?odbc_connect={params}")

try:
    # 2. Load Data
    df = pd.read_csv("orders.csv")

    # 3. Clean existing column names first (replaces spaces with underscores)
    df.columns = [c.lower().replace(' ', '_') for c in df.columns]

    # 4. DRIVE NEW COLUMNS (Calculations)
    df['discount'] = df['list_price'] * df['discount_percent'] * 0.01
    df['sales_price'] = df['list_price'] - df['discount']
    df['revenue'] = df['sales_price'] * df['quantity']
    df['profit'] = df['sales_price'] - df['cost_price']

    # 5. Load to SQL
    # 'replace' will now create a table schema that includes these 4 new columns
    df.to_sql('orders', con=engine, if_exists='replace', index=False, chunksize=50)
    
    print("Success! Data loaded with 4 new calculated columns.")
    print(f"Total columns now: {len(df.columns)}")

except Exception as e:
    print(f"An error occurred: {e}")

finally:
    engine.dispose()
```

```python
import pandas as pd
import sqlalchemy as sa
from urllib.parse import quote_plus

# 1. Connection
params = quote_plus(r"DRIVER={ODBC Driver 17 for SQL Server};SERVER=localhost\SQLEXPRESS;DATABASE=Sales;Trusted_Connection=yes;")
engine = sa.create_engine(f"mssql+pyodbc:///?odbc_connect={params}")

try:
    df = pd.read_csv("orders.csv")
    df.columns = [c.lower().replace(' ', '_') for c in df.columns]
    
    # 2. Prep and Round
    df['discount'] = (df['list_price'] * df['discount_percent'] * 0.01).round(2)
    df['sale_price'] = (df['list_price'] - df['discount']).round(2)
    df['revenue'] = (df['sale_price'] * df['quantity']).round(2)
    df['profit'] = (df['sale_price'] - df['cost_price']).round(2)

    create_table_sql = """
    IF OBJECT_ID('orders', 'U') IS NOT NULL DROP TABLE orders;
    CREATE TABLE orders (
        [order_id] int primary key, [order_date] date, [ship_mode] varchar (20),
        [segment] varchar (20), [country] varchar (20), [city] varchar (20),
        [state] varchar (20), [postal_code] varchar (20), [region] varchar (20),
        [category] varchar (20), [sub_category] varchar (20), [product_id] varchar (50),
        [cost_price] int, [list_price] int, [quantity] int, [discount_percent] int,
        [discount] decimal (20,2), [sale_price] decimal (20,2), [revenue] decimal (20,2), [profit] decimal (20, 2)
    );
    """

    # 3. Use the connection inside the block
    with engine.begin() as conn:
        print("Creating table...")
        conn.execute(sa.text(create_table_sql))
        
        print("Inserting data...")
        df.to_sql('orders', con=conn, if_exists='append', index=False)
        
    print("ðŸš€ Done! Connection closed automatically after success.")

except Exception as e:
    print(f"âŒ Error: {e}")
```
